```python
# Extract systolic and diastolic values
df[['Systolic', 'Diastolic']] = df['Blood Pressure'].str.split('/', expand=True)

# Convert values to integers
df['Systolic'] = df['Systolic'].astype(int)
df['Diastolic'] = df['Diastolic'].astype(int)

# Define categorization thresholds
high_threshold = 130  # Example threshold for systolic
low_threshold = 90   # Example threshold for diastolic

# Create a new column indicating blood pressure category
def categorize_blood_pressure(row):
    if row['Systolic'] >= high_threshold or row['Diastolic'] >= low_threshold:
        return 'High Pressure'
    elif row['Systolic'] < low_threshold and row['Diastolic'] < low_threshold:
        return 'Low Pressure'
    else:
        return 'Normal'

df['Pressure_Category'] = df.apply(categorize_blood_pressure, axis=1)

# Display the resulting DataFrame
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```



Random Forest


```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100,random_state=0)
random_forest_model=clf.fit(X_train, y_train)
y_pred=random_forest_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report
```

```python


print(classification_report(y_test, y_pred))

```
